```python


# keys for users_dictionary
keys=['name','age','work','skills','degree','salary']

# user dictionary
users_dictionary = {}

# skills list
skills_list=[]

# degree_dictionary
degree_dictionary = {}

# degree keys_list
degree_keys = ['title','major','completionDate']

# dictionary_lists
dictionary_list = []

users=0

for total_user in range(int(input("How many users you want to add? "))):
    print("\nUser "+str(users+1))
    for x in range(len(keys)):
        # getting age in integer
        if keys[x] == 'age':
            users_dictionary[keys[x]] = int(input("Enter " + keys[x] + ": "))
        # getting skills in a list
        elif keys[x] == 'skills':
            total_skills = int(input("How many skills you have? "))
            for skills in range(total_skills):
                skills_list.insert(skills,input("Enter skill " + str(skills+1) + ": "))
            users_dictionary[keys[x]] = skills_list
            skills_list=[]
        # getting degree in a dictionary
        elif keys[x] == 'degree':
            for degree_no in range(len(degree_keys)):
                degree_dictionary[degree_keys[degree_no]] = input("Enter degree " + degree_keys[degree_no] + ": ")
            users_dictionary[keys[x]] = degree_dictionary
            degree_dictionary={}
        # getting salary in integer
        elif keys[x] == 'salary':
            users_dictionary[keys[x]] = int(input("Enter " + keys[x] + ": "))
        # saving other keys and values
        else:
            users_dictionary[keys[x]] = input("Enter " + keys[x] + ": ")
    dictionary_list.insert(users,users_dictionary)
    users_dictionary={}
    users +=1
```
